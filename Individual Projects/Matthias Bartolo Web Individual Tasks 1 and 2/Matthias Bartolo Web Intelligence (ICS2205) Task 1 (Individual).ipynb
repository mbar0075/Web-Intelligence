**Which graph-based technique/s are mentioned? describe briefly one of these techniques (if more than one technique is mentioned/used discuss the most important one)**

<br>
Throughout the study, several graph-based techniques were mentioned some of which, include Knowledge graphs, KL-divergence, and Graph embeddings. However, the predominant technique applied was the Knowledge graph data structure which depicted the connections between similar food sources.

A Knowledge graph (KG) is known as a semantic network, which consists of various real-world objects connected through different types of relationships, to represent a network of information usually saved in a graph database [7]. The main components of the KG include:
<br>

1.	**Nodes:** which represent real-world entities

2.	**Edges:** which represent relationships between a pair of nodes

3.	**Labels:** which capture the name of the relationship between nodes


The construction of KG can be facilitated using Ontologies. Ontologies, provide the construction of a conceptual and abstract model of the knowledge base, by classifying and defining general objects which have similar attributes together. Moreover, in ontologies, individual objects which share a partial commonality with other general objects, are identified as subsumees/’ child of’ of the latter objects [8]. 

The classes and relationships identified in the ontologies model would greatly support the construction of a KG, which pictorially depicts the database information in an organised way [8]. An example of the KG data structure can be seen in Figure 3.




 <center><b>  Figure 3: Knowledge Graph data structure (taken from [1]) </b> </center>


Additionally, a characteristic of a KG data structure entails that it is dynamic, meaning that if new inferences are established with new data, the graph will restructure itself [8]. Unfortunately, a drawback of such data structure also exists, and relates to the fact that Knowledge graphs fail to highlight negative instances [9].
